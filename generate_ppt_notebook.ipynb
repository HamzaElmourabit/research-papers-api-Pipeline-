PARTIE 2: TECHNOLOGIES (DÉTAIL COMPLET POUR CHACUNE)
────────────────────────────────────────────────────

SLIDE 2.1: PYTHON 3.13.5
• Langage principal du projet
• Choisi pour: développement rapide, écosystème data science riche
• Utilisation: Scripts ETL, validation données, orchestration
• Librairies clés: pandas, pydantic, cassandra-driver, requests

SLIDE 2.2: DAGSTER (ORCHESTRATION)
• Définition: Plateforme d'orchestration moderne pour data pipelines
• Pourquoi Dagster?
  - Orchestration déclarative vs bash scripts
  - Monitoring et alertes built-in
  - Gestion d'erreurs sophistiquée
  - UI interactive (Dagit) pour visualiser workflows
  - Assets-oriented (pensée par données, pas par tâches)
• Architecture:
  - Assets: fetch_arxiv_papers, validate_papers, store_in_cassandra
  - Jobs: Regroupement d'assets en workflows
  - Resources: Connexions (API ArXiv, Cassandra)
  - Sensors: Triggers automatiques
• Avantages au projet:
  - Chaque étape loggée et tracée
  - Gestion du batch_id (idempotent inserts)
  - Retries automatiques en cas d'erreur

SLIDE 2.3: CASSANDRA (BASE DE DONNÉES NoSQL)
• Définition: Base de données NoSQL distribuée, hautement scalable
• Architecture:
  - Keyspace: arxiv (namespace)
  - Table: papers_raw
  - 13 colonnes: batch_id, arxiv_id, title, abstract, authors, 
                 categories, published_date, updated_date, pdf_url, 
                 json_metadata, ingestion_date, processing_status, notes
• Choix de Cassandra:
  - Scalabilité: Support de milliards de records
  - Performance: Writes ultra-rapides
  - Distributed: Pas de single point of failure
  - Perfect pour Time-Series (papers + timestamps)
• Déploiement: Docker container (cassandra:5.0)
• Schema: CQL (Cassandra Query Language)

SLIDE 2.4: DATABRICKS (PLATEFORME SPARK)
• Définition: Plateforme cloud unifiée pour data engineering + ML + analytics
• Composants utilisés:
  - Spark: Processing distribuée des données
  - Delta Lake: Format de stockage optimisé (ACID transactions)
  - Notebooks: Jupyter-like interface pour transformation
  - SQL Connector: Lecture directe depuis Cassandra
• Couches de données (Medallion Architecture):
  - BRONZE: Ingestion brute depuis Cassandra
  - SILVER: Nettoyage, normalisation, déduplication
  - GOLD: Tables analytiques optimisées pour BI
• Avantages:
  - Transformation distribuée sur gros volumes
  - Features engineering automatisé
  - Scalabilité automatique

SLIDE 2.5: DOCKER (CONTAINERIZATION)
• Définition: Technologie de containerization pour déploiement reproductible
• Utilisation au projet:
  - Cassandra en container (isolé, portable)
  - Reproducibilité: même environnement partout
  - Volumes mappés: Scripts CQL persistés
• Docker Compose: Orchestration multi-containers
• Avantage: "Works on my machine" → "Works everywhere"

SLIDE 2.6: API REST (ArXiv API)
• Source de données: arXiv e-prints (https://arxiv.org)
• Endpoints utilisés:
  - /query?search_query=cat:cs.AI → Fetch papers catégorie AI
  - Support: 5 catégories (CS.AI, CS.LG, CS.CV, CS.CL, stat.ML)
• Rate limiting: ~100 requêtes/heure
• Format: Atom feed → JSON parsing
• Métadonnées extraites: title, abstract, authors, categories, dates

PARTIE 4: ARCHITECTURE DU PROJET

SLIDE 4.1: ARCHITECTURE GLOBALE (DIAGRAMME)

SLIDE 4.2: PHASE 1-4 DAGSTER (ETL PIPELINE)
• Objectif: Extraire, valider, charger 18 articles test
• Processus:
  1. FETCH: Télécharge articles ArXiv (5 catégories)
     - Asset: fetch_arxiv_papers
     - Input: Categories (list)
     - Output: JSON records
  2. VALIDATE: Contrôle qualité Pydantic
     - Asset: validate_papers
     - Rules: Required fields, format validation
     - Output: Pydantic-validated records
  3. STORE: Insertion idempotent en Cassandra
     - Asset: store_in_cassandra
     - Deduplication: batch_id + arxiv_id
     - Output: 18 papers en base
• Status: ✅ 100% COMPLET

SLIDE 4.3: PHASE 5-6 DATABRICKS (ELT AVANCÉ)
• Objectif: Transformation avancée + analytics tables

  - GOLD: 8 tables analytiques

    * fact_metrics, agg_by_category, agg_by_year
